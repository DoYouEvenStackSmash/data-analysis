```python
import numpy as np
from bisect import bisect_left


def binary_search_insert_point(row, target):
    # Binary search to find the insertion point
    return bisect_left(row, target)


def find_insert_points(matrix, vector):
    # Ensure that matrix and vector have compatible shapes
    assert matrix.shape[0] == vector.shape[0]

    # Initialize an array to store insertion points
    insert_points = np.zeros_like(vector, dtype=int)

    # Iterate through each row of the matrix and corresponding element of the vector
    for i, (row, target) in enumerate(zip(matrix, vector)):
        insert_point = binary_search_insert_point(row, target)
        insert_points[i] = insert_point

    return insert_points


insert_points = find_insert_points(s2, np.array(first_column_values))
print(insert_points)
np.save("greedy_indices.npy", [first_column_values, insert_points])
scipy.io.savemat(
    "greedy_indices.mat", {"data": first_column_values, "idx": insert_points}
)
```

```python
import csv
import numpy as np


def read_csv_skip_rows(file_path, skip_rows):
    columns = {}

    with open(file_path, "r") as csv_file:
        csv_reader = csv.reader(csv_file)

        # Skip the first N rows
        for _ in range(skip_rows):
            next(csv_reader)

        # Read the header row
        header = next(csv_reader)

        # Initialize empty lists for each column
        for col in header:
            columns[col] = []

        # Read data into corresponding columns
        for row in csv_reader:
            for col, value in zip(header, row):
                columns[col].append(value)

    # Convert lists to numpy arrays
    for col in columns:
        columns[col] = np.array(columns[col])

    return columns


# Example usage:
file_path = "foo_likelihoods.csv"
skip_rows = 0  # Change this to the number of rows you want to skip

result = read_csv_skip_rows(file_path, skip_rows)

# Access the arrays for each column using column names
for col_name, col_array in result.items():
    print(f"{col_name}: {col_array}")
```
